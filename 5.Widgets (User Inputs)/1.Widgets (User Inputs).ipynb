## üéõÔ∏è **Widgets (User Inputs)** ‚Äì *Streamlit v1.47.0*

Streamlit widgets allow users to **interact** with your app in real time. Each widget triggers a **re-run** of your script upon interaction.

---

### üîò `st.button(label, key=None, help=None, on_click=None, args=None, kwargs=None)`

* **Purpose:** Triggers an action when clicked.
* **Returns:** `True` on click.
* ‚úÖ Use when: You need user-driven triggers (e.g. search, submit).

```python
if st.button("Greet"):
    st.success("Hello there!")
```

---

### ‚òëÔ∏è `st.checkbox(label, value=False, key=None, help=None)`

* **Purpose:** Boolean on/off switch.
* ‚úÖ Use when: Enabling optional features or filters.

```python
if st.checkbox("Show data"):
    st.dataframe(data)
```

---

### üîò `st.radio(label, options, index=0, ...)`

* **Purpose:** Select one from multiple options.
* ‚úÖ Use when: Only one choice should be allowed.

```python
genre = st.radio("Choose genre", ["Comedy", "Drama", "Action"])
```

---

### üîΩ `st.selectbox()` / `st.multiselect()`

| Function           | Purpose                   | Example                                  |
| ------------------ | ------------------------- | ---------------------------------------- |
| `st.selectbox()`   | Single dropdown selection | `fruit = st.selectbox("Fruit", [...])`   |
| `st.multiselect()` | Select multiple options   | `opts = st.multiselect("Choose", [...])` |

---

### üéöÔ∏è `st.slider()` / `st.select_slider()`

* **Purpose:** Select numeric or categorical values using a slider.
* ‚úÖ Use when: Range-based input is needed.

```python
age = st.slider("Select age", 0, 100)
```

```python
color = st.select_slider("Pick", options=["Red", "Green", "Blue"])
```

---

### ‚å®Ô∏è `st.text_input()` / `st.text_area()`

| Function          | Purpose                | Example                                  |
| ----------------- | ---------------------- | ---------------------------------------- |
| `st.text_input()` | Single-line text field | `name = st.text_input("Your name")`      |
| `st.text_area()`  | Multi-line text box    | `msg = st.text_area("Write message...")` |

---

### üî¢ `st.number_input()` / `st.date_input()`

* **Purpose:** Numeric or calendar date inputs.
* ‚úÖ Use when: You need typed or selected values.

```python
qty = st.number_input("Quantity", min_value=1)
date = st.date_input("Pick a date")
```

---

### üìÅ `st.file_uploader(label, type=None, ...)`

* **Purpose:** Upload files (CSV, images, audio, etc.)
* ‚úÖ Use when: You want users to input files into the app.

```python
csv = st.file_uploader("Upload CSV", type="csv")
```

---

### üì∑ `st.camera_input(label)`

* **Purpose:** Take a photo using the user's camera.
* ‚úÖ Use when: You want quick image capture.

```python
img = st.camera_input("Take a picture")
```

---

### ü™Ñ `st.toggle(label, value=False, ...)` *(New in v1.47.0)*

* **Purpose:** A sleek boolean switch (UX upgrade over `st.checkbox`)
* ‚úÖ Use when: You want a more modern toggle UI.

```python
dark_mode = st.toggle("Dark Mode")
```
